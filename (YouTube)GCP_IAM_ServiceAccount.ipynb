```python
from google.oauth2 import service_account
from google.cloud import bigquery

PROJECT_ID = 'hyunsoo-it'
LOCATION = 'asia-northeast3'
JSON_KEY_PATH = '/content/bq-test-key.json'

# 서비스 계정 키 파일을 사용하여 인증 객체 생성
credentials = service_account.Credentials.from_service_account_file(JSON_KEY_PATH)


client = bigquery.Client(
    project=PROJECT_ID,
    credentials=credentials,
    location=LOCATION
)
```

```python
## dataset 목록 조회

for dataset in client.list_datasets():
    print(dataset.dataset_id)
```

```python
## 테이블 목록 조회

for table in client.list_tables(dataset='bq_data'):
    print(table.table_id)
```

```python
## dataset 생성

dataset_name = 'hyunsoo_dataset'

client.create_dataset(
    dataset=f"{PROJECT_ID}.{dataset_name}",
    exists_ok=True
)
```

```python
## dataset 삭제

client.delete_dataset(
    dataset=f"{PROJECT_ID}.{dataset_name}",
    not_found_ok=True
)
```

```python
## 테이블 데이터 조회 (BigQuery 테이블 -> Pandas Dataframe)

query_obj = client.query(
    query="SELECT * FROM bq_data.sales"
)

df = query_obj.to_dataframe()

df.head()
```
